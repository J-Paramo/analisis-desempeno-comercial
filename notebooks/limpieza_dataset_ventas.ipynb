```python
# Liberias necesarias
import pandas as pd
import numpy as np

# Cargar dataset ventas
url_ventas = '../data/raw/dataset_ventas_2025_09.csv'
df = pd.read_csv(url_ventas)
#df.head()

# Cargar dataset empleados
url_empleados = '../data/raw/dataset_empleados_alcom.csv'
df_empleados = pd.read_csv(url_empleados)
df_empleados.head()
```

```python
# Ver columnas y tipo de datos
#df.info()
#df_empleados.info()

# Renombrar columnas
df.rename(columns = {
    'Precio_Mensual_del_paquete_formado': 'precio_paquete',
    'paquete_vendido': 'paquete',
    'fechaventa': 'fecha'
}, inplace=True)

df_empleados.rename(columns = {
    'emp_Nroemp': 'empleado',
    'emp_nombre': 'nombre_empleado'
}, inplace=True)

# Cambiar el tipo de dato de fecha y extraer solo la fecha (sin hora)
df.loc[:, 'fecha'] = pd.to_datetime(df['fecha']).dt.date
# Buscar valores nulos
#df.isnull().sum()
df.isnull().sum()
```

```python
#df.head()
#df_empleados.head()

# Merge del nombre del empleado
df = df.merge(
    df_empleados[['empleado', 'nombre_empleado']],
    left_on='agente',
    right_on='empleado',
    how='left'
)
df.rename(columns={'nombre_empleado': 'nombre_agente'}, inplace=True)
df.drop(columns=['empleado'], inplace=True)

# Merge del nombre del supervisor
df = df.merge(
    df_empleados[['empleado', 'nombre_empleado']],
    left_on='supervisor',
    right_on='empleado',
    how='left'
)
df.rename(columns={'nombre_empleado': 'nombre_supervisor'}, inplace=True)
df.drop(columns=['empleado'], inplace=True)

df.head()
# Eliminar las filas con valores nulos o vacios
df = df.dropna(subset=['paquete', 'precio_paquete', 'nombre_agente', 'nombre_supervisor'])
```

```python
# Guardar archivo limpio
df.to_csv('../data/processed/dataset_limpio.csv', index=False)
```
